```python
# find all tree branches for a given cluster
def find_branches(cluster, tree, cluster_labels):
    print(cluster_labels[cluster])
    branches = []
    for branch, subbranches in tree.items():
        if isinstance(subbranches, dict):
            for subbranch, clusters in subbranches.items():
                if cluster in clusters:
                    branches.append((branch, subbranch))
        else:
            if cluster in subbranches:
                branches.append(branch)
    print(branches)
    return branches


branches = find_branches(3, cluster_tree, cluster_labels)
```
