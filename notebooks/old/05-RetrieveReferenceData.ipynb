# Get Reference Data from the APIs


```python
import pandas as pd

from src.data.ReferenceDataFetcherFromDoi import RefsFetcher
```

```python
df = pd.read_pickle("../data/processed/merged/merged_data.pkl")
df.shape
```

```python
# create chunks of 2000
df_chunks = [df[i : i + 2000] for i in range(0, df.shape[0], 2000)]

for i in range(len(df_chunks)):
    refs_fetcher = RefsFetcher(df_chunks[i])
    refs_fetcher.reference_fetcher()
    refs_fetcher.calculate_and_print_avg()
    df_chunks[i] = refs_fetcher.df
    refs_fetcher.df.to_pickle(
        "../data/processed/merged/merged_data_with_refs_" + str(i) + ".pkl"
    )

df = pd.concat(df_chunks)

df.to_pickle("../data/processed/merged/merged_data_with_refs.pkl")
```
