```python
import re
import sys

sys.path.append("02_AllThingsData/Utils/DataRetrieval/WHO-ATC-Scraper")

from WHO_ATC_Scraper import ATCScraper


def add_drug_class(df, drug_column="Drug", class_column="Class"):
    """
    Fetches ATC drug classification data, classifies the drugs in the given DataFrame,
    and inserts a new column with the drug class information.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing drug data.
    - drug_column (str): The column name in df where drug names are stored.
    - class_column (str): The desired name of the new column to be inserted into df with drug class information.
    """

    # Instantiate ATCScraper, fetch and scrape data
    atc_scraper = ATCScraper()
    atc_scraper.fetch_data()
    atc_scraper.scrape_and_store_data()

    # Classify drugs in df and store the results in a list
    classes = []
    for drug in df[drug_column]:
        drug_class = None
        for key, value in atc_scraper.dataframes_dict.items():
            if drug in value["Name"].to_list():
                drug_class = key
        if drug_class is None:
            classes.append("parent class")
        else:
            classes.append(drug_class)

    # Insert new column with drug class information into df
    df.insert(1, class_column, classes)
    return df


df = add_drug_class(df)
```

# Prepare the data for the plot


## load color


```python
# load color data
df_color = pd.read_csv("02_AllThingsData/Utils/Plotting/Data/ColorPalettes_PerATC.csv")
df_color["DrugClass"] = df_color["DrugClass"].str.lower().str.replace("-", " ")
```

```python
def assign_colors_to_drugs(df: pd.DataFrame, df_color: pd.DataFrame) -> pd.DataFrame:
    """
    Assigns colors to drugs in a DataFrame based on matching DrugName or DrugClass in a reference DataFrame. When a parent class is matched, the associated color palette is chosen (e.g. Blues) and used as e.g. blue.

    Parameters:
    df (pd.DataFrame): The input DataFrame containing a column "Drug" with drug names.
    df_color (pd.DataFrame): The reference DataFrame containing columns "DrugName", "DrugClass", and associated colors.

    Returns:
    pd.DataFrame: The input DataFrame with an additional column "Color" indicating the assigned color for each drug.
    """
    colors = []
    for drug in df["Drug"]:
        color = "black"  # Default to "black"
        if drug in df_color["DrugName"].values:
            color = df_color[df_color["DrugName"] == drug]["Color"].values[0]
        elif drug in df_color["DrugClass"].values:
            color = (
                df_color[df_color["DrugClass"] == drug]["ColorPalette"]
                .values[0]
                .lower()[:-1]
            )  # You might want to choose the first color, using .values[0]

        colors.append(color)  # Add the determined color to the list

    df["Color"] = colors  # Add the colors list as a new column in the DataFrame
    return df


# Usage Example:
# You need to have your df and df_color defined and loaded with your data before calling the function
df = assign_colors_to_drugs(df, df_color)
df.head(5)
```
