```python
params = "alpha0.3_k20_res0.005"
p = f"../data/06-clustered-df/{params}.pkl"
df = pd.read_pickle(p)

# print(df.columns)

df_dict = {}
for cluster in df["cluster_alpha0.3_k20_res0.005"].unique():
    # create df to key and sort by centrality
    df_dict[cluster] = df[df["cluster_alpha0.3_k20_res0.005"] == cluster].sort_values(
        by="centrality_alpha0.3_k20_res0.005", ascending=False
    )
```

```python
def cluster_title_print(df_dict, cluster, random=False, n=10):
    if random:
        titles = df_dict[cluster].sample(n, random_state=1887)["title"].values
    else:
        titles = df_dict[cluster].head(n)["title"].values
    print(f"Cluster {cluster}")
    for title in titles:
        print(title)
        print("# # " * 30)
    print("\n")


cluster_title_print(df_dict, 91, random=True, n=20)
```
