```python
# Assuming df_summary is your DataFrame and it has columns named "Cluster" and "Word_0"
def cluster_number_to_label(df_summary, cluster_numbers: list):
    clusters = df_summary["Cluster"].tolist()
    merged_words = [
        str(row["Word_0"]) + "; " + str(row["Word_1"])
        for index, row in df_summary.iterrows()
    ]
    clusters_words_dict = dict(zip(clusters, merged_words))
    cluster_number_label_dict = {k: clusters_words_dict[k] for k in cluster_numbers}
    return cluster_number_label_dict


clusters_to_plot = [34, 67, 5]
cluster_number_label_dict = cluster_number_to_label(df_summary, clusters_to_plot)
cluster_number_label_dict
```

```python
side_effect_clusters = {
    3: "pulmonary hypterension",
    65: "gastrointestinal bleeding",
    37: "serotonin syndrome",
    42: "ssri induced mania",
    51: "sexual dysfunction",
    60: "Hyponatremia",
    # 62: "glaucoma",
    # 79: "suicide",
    # 80: "ssri induced apathy",
    # 105: "discontinuation / withdrawal",
    # 109: "risk of fracture",
}

import matplotlib.pyplot as plt


def plot_rolling_cluster_distribution(df, cluster_column, clusters_labels_dict):
    clusters_list = list(clusters_labels_dict.keys())
    df_sub = df[df[cluster_column].isin(clusters_list)]

    # Set the size of the figure
    plt.figure(figsize=(15, 10))

    # Group by year and cluster, then calculate normalized value counts
    # Including the group_keys=True parameter as per the warning suggestion
    cluster_distribution = (
        df_sub.groupby(["year", cluster_column], group_keys=True)
        .size()
        .groupby(level=0)
        .apply(lambda x: x / x.sum())
        .unstack()
    )

    # Calculate the rolling mean for smoothing
    rolling_cluster_distribution = cluster_distribution.rolling(window=2).mean()

    # Plot the data
    for cluster in clusters_list:
        plt.plot(
            rolling_cluster_distribution[cluster], label=clusters_labels_dict[cluster]
        )

    # Setting the legend outside the plot
    plt.legend(title="Cluster Labels", loc="upper left", bbox_to_anchor=(1, 1))

    plt.title("Cluster Distribution Over Time (Rolling Mean)")
    plt.xlabel("Year")
    plt.ylabel("Normalized Cluster Count")

    # Display the plot
    plt.show()


plot_rolling_cluster_distribution(df_0025, "cluster", side_effect_clusters)
```
